```python
with open('demo.txt','w')as file:
  file.write("hello  it's a demo file ")
with open('demo.txt','r')as file:
  fo = file.read()
  print(fo)
```

```python
with open('demo.txt','w')as file:
  file.write("this is first line\n its second line \n its third line")
with  open('demo.txt', 'r')as file:
  l1 = file.readline()
  print(l1)
  l2 = file.readline()
  print(l2)
  l3 = file.readline()
  print(l3)
```

```python
from __future__ import print_function
with open('demo.txt','a')as file:
  file.write("thhis is 4th line,,,,")
with open('demo.txt')as file:
    a = file.read()
    print(a)
```

```python
try:
  with open('demo.txt','x')as file:
    file.write("hello")
except FileExistsError as e:
  print(e)
finally:
  with open('demo.txt','r')as file:
    a=file.read()
    print(a)
```
